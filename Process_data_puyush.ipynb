```python
import pandas
pandas.set_option('display.max_rows',1000)
pandas.set_option('display.max_columns',10000)
import collections
```

```python
m2016 = pandas.read_csv('music2016.csv')
m2016 = m2016.drop(m2016.columns[0], axis=1)

m2016['chord_set'] = [set(chords.split(' ')) for chords in m2016.chords]
cd_2016_set = set().union(*m2016.chord_set)

# counts the number of chords for each row and makes a pair (cord, count) into a set
def term_feq_list(chords): 
    li = list()
    cd0_list = chords.split(' ')
    for x in cd0_list:
        t = tuple()
        val = cd0_list.count(x)
        t += (x, val)
        li.append(t)
    return li
#  for the pairs for each rows chords
def pairs_of_chords(chord_row):
    chord_pair = list()      
    cd_list = chord_row.split(' ')
    for index in range(0,len(cd_list)-1):
        temp_str = str(cd_list[index] + ' ' + cd_list[index+1])
        chord_pair.append(temp_str)
    return chord_pair
def term_feq_pairs(chords): 
    li = list()
    chords = str(chords)
    cd0_list = chords.split(',')
    for x in cd0_list:
        t = tuple()
        val = cd0_list.count(x)
        t += (x, val)
        li.append(t)
    return li


m2016['term_feq_list'] = [set(term_feq_list(chords))for chords in m2016.chords]
m2016['pairs_of_chords'] = [pairs_of_chords(chords_row)for chords_row in m2016.chords]
m2016['term_feq_pairs'] = [set(term_feq_pairs(chords))for chords in m2016.pairs_of_chords]
m2016
```
